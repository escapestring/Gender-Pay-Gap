```python
#select needed variables
f1 = f1[["employee_id", "gender", "annual_salary_amount_in_usd", "job_experience", "job_pay_level", "full_or_part_time", "hourly_or_salary", "job_family_group", "job_continent"]]

f1.head()
```

```python
#replace comma in numeric value (annual_salary_amount_in_usd)
f1["annual_salary_amount_in_usd"] = f1["annual_salary_amount_in_usd"].str.replace(",", "")

f1.head(5) #view top 5 of dataframe
```

```python
#convert numeric string values into numeric data type
f1["annual_salary_amount_in_usd"] = f1["annual_salary_amount_in_usd"].apply(pd.to_numeric)

f1.info()
```

```python
# Convert the job_experience column to integers, handling NaN values
f1['job_experience'] = pd.to_numeric(f1['job_experience'], errors='coerce').astype('Int64')

f1.info() #see data type of variables
```

```python
#view data
f1.head()
```

```python
# Counting the number of missing values in each column
missing_values = f1.isnull().sum()

# Creating a new DataFrame to display missing values
f1_missing = pd.DataFrame(missing_values, columns=['Missing Values'])

f1_missing #show missing values by columns
```

```python
#check unique values of job experience
f1["job_experience"].unique()
```

```python
#remove all missing values
f1 = f1.dropna()

f1.shape
```

```python
# filter dataframe to have only male and female gender and no NAN values
q = """SELECT *
        FROM f1
        WHERE gender in ('Male', 'Female')"""

f1 = pysqldf(q)   #pass query into function to filter dataframe
f1 #view dataframe
```

```python
# Function to process each string and remove (Job Family Group)
import re

def reformat_string(text):
    # This regex matches a space followed by anything inside parentheses
    return re.sub(r'\s+\([^)]*\)', '', text)
```

```python
# Apply the function to each element in the column
f1['job_family_group'] = f1['job_family_group'].apply(reformat_string)

f1.head()
```
